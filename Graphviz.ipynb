```python
from graphviz import Digraph
```

```python
# 간단한 방향성 그래프 생성

dot = Digraph(comment='한국의 주요 도시')
dot.attr(rankdir='LR') # 가로 방향으로 그래프 설정
dot.attr('node', fontname='NanumGothic')
dot.attr('edge', fontname='NanumGothic')

# 노드 추가
dot.node("A", "서울")
dot.node("B", "부산")
dot.node("C", "대전")

# 간선 추가
dot.edge("A", "B", label="경부선")
dot.edge("A", "C", label="호남선")
dot.edge("B", "C", label='동해선')

# 그래프 출력
print(dot.source)

# 이미지 파일로 저장 후 자동으로 열기
dot.render('test-output/korean-cities.gv', view=True)
```
