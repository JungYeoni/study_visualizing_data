```python
import numpy as np
from bokeh.io import output_notebook, show
from bokeh.plotting import figure

output_notebook()
```

```python
# 산포도 - relation
p = figure(width=700, height=400)
```

```python
show(p) # 아직 아무것도 없기 때문에 비어있음
```

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 2, 3, 4, 5]

p.circle(x, y, size=10)

show(p)
```

```python
# 원에 외곽선이랑 채우는 색 선택
p.circle(x, y, size=10, line_color='orange', fill_color='navy')
show(p)
```

```python
# 원에 외곽선이랑 채우는 색 선택
p.circle(x, y, size=30, line_color='orange', fill_color='brown')
show(p)
```

```python
# 투명도
p.circle(x, y, size=30, line_color='orange', fill_color='blue', alpha=0.1)
show(p)
```

```python
# 필요한 라이브러리 임포트
import bokeh
import bokeh.sampledata
import pandas as pd
from bokeh.plotting import figure
from bokeh.sampledata.stocks import AAPL
```

```python
bokeh.__version__
```

```python
df = pd.DataFrame(AAPL)
df.head()
```

```python
pd.to_datetime(df['date'])
df = df.drop('date', axis=1)
```

```python
p = figure(width=700, height=300, x_axis_type='datetime')
p.line(df.index, df['close'], color='navy', alpha=0.5)
show(p)
```

```python
from re import template
# 웹으로 내보내는데 필요한 라이브러리 임포트
import jinja2
from bokeh.embed import components

template = jinja2.Template("""
<html>
<head>
  <link href="https://cdn.pydata.org/bokeh/release/bokeh-3.7.3.min.css" rel="stylesheet" type="text/css">
  <script src="https://cdn.pydata.org/bokeh/release/bokeh-3.7.3.min.js"></script>
  <title>bokeh test</title>
</head>

<body>
  <h1>This is site for Bokeh Test</h1>
  이 사이트는 보케 테스트 하기 위한 공간 입니다.

  {{ script }}
  {{ div }}

</body>
</html>
""")
```

```python
script, div = components(p)

from IPython.display import HTML
HTML(template.render(script=script, div=div))
```

```python
with open('a.html', 'w') as f:
  f.write(template.render(script=script, div=div))

import webbrowser

webbrowser.open('a.html');
```
