```python
# 라이브러리 임포트
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='darkgrid') # style에 따라 다르게 나타남

%matplotlib inline
```

```python
# 데이터셋
tips = sns.load_dataset('tips')
tips
```

```python
sns.relplot(x='total_bill', y='tip', data=tips)
```

- `relplot`을 그냥 쓰면 분포를 나타내는 것을 볼 수 있음
- 현재 그래프에서는 우상향하는 것을 확인할 수 있음

```python
# 색상을 넣어서 비흡연자 vs 흡연자 비교
sns.relplot(x='total_bill', y='tip', data=tips, hue='smoker')
```

- `smoker`은 범주형 데이터
- `hue`는 이에 따라 색상을 다르게 표시하는 데 사용

```python
# hue 색상
sns.relplot(x='total_bill', y='tip', data=tips, hue='size')
```

```python
# style 적용
sns.relplot(x='total_bill', y='tip', data=tips, hue='size', style='smoker')
```

- 색상이 보편적으로 확연한 시각적 효과를 줄 수 있음

```python
# 색상 선택
sns.relplot(x='total_bill', y='tip', data=tips, hue='size', palette='Blues')
```

```python
# 사이즈
sns.relplot(x='total_bill', y='tip', data=tips, size='size')
```

```python
fmri = sns.load_dataset('fmri')
fmri
```

```python
sns.relplot(x='timepoint', y='signal', kind='line', data=fmri) # default는 표준편차
```

- 위아래로 오차 범위를 표시해준것

```python
sns.relplot(x='timepoint', y='signal', kind='line', data=fmri, ci=None)
```

`ci=None` 하면 오차범위 표시x

```python
sns.relplot(x='timepoint', y='signal', kind='line', data=fmri, ci='sd') # 분산으로 보여주는 것
```

```python
sns.relplot(x='timepoint', y='signal', kind='line', data=fmri,hue='region', style='event')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='point')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='swarm', hue='sex')
```

### `sns.catplot` 상세 설명

`catplot`은 **Figure-level** 인터페이스로, 여러 가지 범주형 그래프를 그릴 수 있게 해줍니다. 내부적으로 `FacetGrid`를 사용하여 서브플롯을 생성하기에 용이합니다.

#### 주요 매개변수:
- `x`, `y`: 축 데이터
- `hue`: 색상을 구분할 범주형 변수
- `data`: 데이터프레임
- `kind`: 그래프의 종류 지정
- `col`, `row`: 데이터를 나누어 그릴 기준 컬럼

### 그래프 합치기 - jointplot

```python
# 데이터 만들기
df = pd.DataFrame(np.random.randn(500, 2), columns=[1, 2])
df
```

```python
sns.relplot(x=1, y=2, data=df)
```

```python
sns.jointplot(x=1, y=2, data=df)
```

- 정규화된 모양을 출력하고 있음을 확인할 수 있음

```python
# 다른 모양 - kind를 수정하면 됨
sns.jointplot(x=1, y=2, data=df, kind='hex')
```

```python
sns.jointplot(x=1, y=2, data=df, kind='kde')
```

```python
sns.jointplot(x=1, y=2, data=df, kind='resid') # 투명도를 반영함
```

```python
sns.jointplot(x=1, y=2, data=df, kind='reg')
```
