```python
import chartify
import pandas as pd
```

```python
# 리셋
from bokeh.io import reset_output, output_notebook

reset_output()
output_notebook()
```

```python
# 예제 데이터 생성
df = pd.DataFrame({
    'x': ['A', 'B', 'C'],
    'y': [10, 20, 15]
})

# chartify 객체 생성
ch = chartify.Chart(blank_labels = True, x_axis_type='categorical')
ch.set_title("Sample Bar Graph")
ch.set_subtitle("This is a simple bar graph example")
ch.plot.bar(
    data_frame=df,
    categorical_columns='x',
    numeric_column='y'
    )

# 차트 생성
ch.show()
```

```python
# HTML로 저장
ch.save("chart.html")
```
