```python
import folium
```

```python
# 마커 그리기
m = folium.Map(location=[37.503, 126.770], zoom_start=16)
tooltip = 'Chilk me!'

folium.Marker([37.503, 126.770], popup='<i>위</i>', tooltip=tooltip).add_to(m)
folium.Marker([37.50, 126.770], popup='<i>가운데</i>', tooltip=tooltip).add_to(m)
folium.Marker([37.497, 126.770], popup='<i>아래</i>', tooltip=tooltip).add_to(m)
m
```

```python
# 마커 변경하기
m = folium.Map(location=[37.503, 126.770], zoom_start=16)

# icon && color
folium.Marker([37.503, 126.770], icon=folium.Icon(icon='cloud')).add_to(m)
folium.Marker([37.50, 126.770], icon=folium.Icon(color='green')).add_to(m)
folium.Marker([37.497, 126.770], icon=folium.Icon(icon='info-sign', color='red')).add_to(m)

m
```

```python
# icon && color
folium.Marker([37.503, 126.770], icon=folium.Icon(icon='star')).add_to(m)
folium.Marker([37.50, 126.770], icon=folium.Icon(color='green')).add_to(m)
folium.Marker([37.497, 126.770], icon=folium.Icon(icon='bookmark', color='red')).add_to(m)

m
```

```python
m = folium.Map(location=[37.503, 126.770], zoom_start=16)

# icon && color
folium.Marker([37.503, 126.770], icon=folium.Icon(icon='cloud')).add_to(m)
folium.Marker([37.50, 126.770], icon=folium.Icon(color='green')).add_to(m)
folium.Marker([37.497, 126.770], icon=folium.Icon(icon='info-sign', color='red')).add_to(m)
folium.Circle([37.50, 126.773], radius=100, color='#ABF200', fill=True, fill_color='#ABF200').add_to(m)
folium.Circle([37.50, 126.773], radius=200, color='#ABF200', fill=True, fill_color='#ABF200').add_to(m)

m
```

```python
# 지도를 새로 생성하여 이전의 잘못된 설정을 초기화합니다.
m = folium.Map(location=[37.503, 126.770], zoom_start=16)

# 클릭 시 마커를 추가하는 기능을 설정합니다. - extra feature
m.add_child(folium.ClickForMarker(popup='Warning'))
m
```

```python
# 경도 위도를 팝업하는 기능
m = folium.Map(location=[37.503, 126.770], zoom_start=16)

# icon && color
folium.Marker([37.503, 126.770], icon=folium.Icon(icon='cloud')).add_to(m)
folium.Marker([37.50, 126.770], icon=folium.Icon(color='green')).add_to(m)
folium.Marker([37.497, 126.770], icon=folium.Icon(icon='info-sign', color='red')).add_to(m)
folium.Circle([37.50, 126.773], radius=100, color='#ABF200', fill=True, fill_color='#ABF200').add_to(m)
folium.Circle([37.50, 126.773], radius=200, color='#ABF200', fill=True, fill_color='#ABF200').add_to(m)

m.add_child(folium.LatLngPopup())
m
```
